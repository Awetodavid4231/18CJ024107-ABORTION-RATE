```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import statistics
```

```python
countries = [
'Albania',
'Armenia',
'Australia',
'Austria',
'Azerbaijan',
'Bahrain',
'Belarus',
'Belgium',
'Brazil',
'Bulgaria',
'Canada',
'China',
'Croatia',
'Cuba',
'Czech Republic',
'Denmark',
'Dominican Republic',
'Estonia',
'Finland',
'France',
'Georgia',
'Germany',
'Greece',
'Hungary',
'Iceland',
'India',
'Israel',
'Italy',
'Japan',
'Kazakhstan',
'Kyrgyzstan',
'Latvia',
'Lithuania',
'Macedonia',
'Mexico',
'Mongolia',
'Netherlands',
'New Zealand',
'Norway',
'Panama',
'Poland',
'Portugal',
'Qatar',
'Republic of Moldova',
'Romania',
'Russian Federation',
'Seychelles',
'Singapore',
'Slovakia',
'Slovenia',
'South Africa',
'Spain',
'Sweden',
'Switzerland',
'Tajikistan',
'Tunisia',
'Ukraine',
'United Kingdom',
'United States of America',
'Uzbekistan',
'Vietnam',

        ]

df   = pd.read_csv('ABT.csv')
df = df[ (df['Country or Area'].isin(countries) )]
df = df.sort_values('Value')
df = df.set_index('Country or Area')
df
```

```python
for i in range(3):
    l = 1996+i*3
    h = l + 3
    fig, axes  = plt.subplots(1,3, figsize=(5,10), sharey=True)
 
    for idx, year in enumerate(range(l,h)):
        df[df['Year'] == year]['Value'].plot.barh(ax=axes[idx])
        axes[idx].set_title(year)
```

```python
mean_vals = []
for year in range(1996,2005):
    mean_vals.append(df[df['Year'] == year]['Value'].mean())
    
plt.plot(range(1996,2005), mean_vals)
plt.title('Mean for Abortion Rate of countries from 1996-2005')
plt.xlabel('Year')
plt.ylabel('Ratio')
```

```python
stdev_vals = []
for year in range(1996,2005):
    stdev_vals.append(df[df['Year'] == year]['Value'].var()**0.5)
    
plt.plot(range(1996,2005), stdev_vals)
plt.title('Standard Deviation for Abortion Rate of countries from 1996-2005')
plt.xlabel('Year')
plt.ylabel('Ratio')
```

```python
iqr_vals = []
def get_iqr(sample):
    q3,q1 = np.percentile(sample, [75 ,25])
    return q3 - q1

for year in range(1996,2005):
    iqr_vals.append(get_iqr(df[df['Year'] == year]['Value']))
    
    
plt.title('Inter Quartile for Abortion Rate of countries from 1996-2005 ')
plt.xlabel('Year')
plt.ylabel('Ratio')
plt.plot(range(1996,2005), iqr_vals)
```
